```python
def load_data(file_path):
    """
    加载数据
    """
    data = []
    event_types = {}


    with open(file_path) as f:
        for line in f:
            line = line.strip()
            contents = line.split(',"')
            text_id = contents[0][1:][:-1]
            text_content  = contents[1][:-1]
            event_type = contents[2][:-1]
            if event_type not in event_types:
                event_types[event_type] = 0
            event_types[event_type] += 1
            # 如果是eval/test数据集
            if len(contents) == 3:
                data.append([text_id, text_content, event_type])
            else:
                event_subject = contents[3][:-1]
                data.append([text_id, text_content, event_type, event_subject])
                print(event_subject)
                
    return data, event_types
```
